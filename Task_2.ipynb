```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic_data = pd.read_csv(url)
```

```python
print(titanic_data.head())
```

```python
print("Missing values:\n", titanic_data.isnull().sum())
```

```python
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)
```

```python
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
```

```python
titanic_data.drop('Cabin', axis=1, inplace=True)
```

```python
print("Missing values after cleaning:\n", titanic_data.isnull().sum())
```

```python
sns.countplot(x='Pclass', data=titanic_data)
plt.title('Distribution of Passenger Classes')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=titanic_data)
plt.title('Survival Count based on Passenger Class')
plt.show()
```

```python
sns.histplot(titanic_data['Age'], bins=20, kde=True)
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=titanic_data)
plt.title('Survival Count based on Gender')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=titanic_data)
plt.title('Age Distribution based on Survival')
plt.show()
```
