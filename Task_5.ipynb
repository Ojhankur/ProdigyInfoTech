```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
accident_data = pd.read_csv('/content/UK_Accident.csv')
```

```python
sns.countplot(x='Road_Type', data=accident_data)
plt.title('Accidents by Road Condition')
plt.show()
```

```python
plt.scatter(accident_data['Longitude'], accident_data['Latitude'], alpha=0.5)
plt.title('Accident Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
